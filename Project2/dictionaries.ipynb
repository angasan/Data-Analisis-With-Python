```python
import csv
```

```python
def process_csv(filename):
    example_file = open(filename, encoding="utf-8")
    example_reader = csv.reader(example_file)
    example_data = list(example_reader)
    example_file.close()
    return example_data
```

```python
csv_rows = process_csv("vaccinations.csv")
csv_header = csv_rows[0]
csv_data = csv_rows[1:]

print(csv_header)
```

```python
def cell(row_idx, col_name):
    """
    Returns the data value (cell) corresponding to the row index and 
    the column name of a CSV file.
    """
    integers = ["total_vaccinations", "daily_vaccinations", "people_vaccinated", "people_fully_vaccinated",
               "total_boosters","population"]
    col_idx = csv_header.index(col_name)
    val = csv_data[row_idx][col_idx]
    if val == "": 
        return None
    elif col_name in integers:
        val = int(val)
    return val
```

```python
#Q1: What dates are mentioned in the dataset?
dates = list()
for row_idx in range(len(csv_data)):
    date = cell(row_idx, "date")
    dates.append(date)
dates = list(set(dates))
dates
```

```python
#Q2: What countries are listed in the dataset?
names = list()
for row_idx in range(len(csv_data)):
    name = cell(row_idx, "country")
    names.append(name)
names = list(set(names))
names
```

```python
#Q3: What is the population of each country in the dataset?
populations = {}
for row_idx in range(len(csv_data)):
    country = cell(row_idx, "country")
    population = cell(row_idx, "population")
    if country == None:
        continue
    if country not in populations: 
        populations[country] = None
    if population != None:
        populations[country] = population
populations
```

```python
#Q4: How many countries have an entry for daily vaccination on Jan 28, 2022?
daily_vacc = list()
for row_idx in range(len(csv_data)):
    country = cell(row_idx, "country")
    date = cell(row_idx, "date")
    daily_vaccination = cell(row_idx, "daily_vaccinations")
    if date == '01/28/2022' and daily_vaccination != None:
        daily_vacc.append(country)
len(set(daily_vacc))
```

```python
def daily_vaccinations_on(search_date):
    '''return a dictionary mapping each country to
    the number of daily vaccinations on the given date;
    if data is missing, the value for that country is None'''
    dic = {}
    for row_idx in range(len(csv_data)):
        date = cell(row_idx, "date")
        country = cell(row_idx, "country")
        daily_vaccinations = cell(row_idx, "daily_vaccinations")
        if country not in dic:
            dic[country] = None
        if daily_vaccinations != None and date == search_date:
            dic[country] = daily_vaccinations
    return dic
```

```python
#Q5: For each country, how many vaccinations were given on Jan 31, 2022?
daily_vaccinations_on('01/31/2022')
```

```python
#Q6: For each country, what is the sum of the vaccinations given on all of the dates mentioned in the dataset?
sum_of_vaccinations = {}
for row in dates:
    curr_dic = daily_vaccinations_on(row)
    if sum_of_vaccinations == {}:
        sum_of_vaccinations = curr_dic
    else:
        for row in sum_of_vaccinations:
            if curr_dic[row] == None:
                continue
            elif sum_of_vaccinations[row] == None:
                sum_of_vaccinations[row] = curr_dic[row]
            else:
                sum_of_vaccinations[row] += curr_dic[row]
            
sum_of_vaccinations
```

```python
def get_max_key(dic):
    max_value = 0
    max_key = None

    # iterate over the dict to find the max
    for row in dic:
        curr_value = dic[row]
        if curr_value == None:
            continue
        if curr_value > max_value or max_key == None:
            max_value = curr_value
            max_key = row
            
    return max_key
```

```python
#Q7: Which country completed the most number of vaccinations during the dates mentioned in the dataset?
get_max_key(sum_of_vaccinations)
```

```python
#Q8: How many vaccinations (across the world) were given on each date?
dic_dates = {}

for rows in dates:
    count = 0
    dic_spec_date = daily_vaccinations_on(rows)
    for row in dic_spec_date: 
        if dic_spec_date[row] == None:
            continue
        count += dic_spec_date[row]
    dic_dates[rows] = count
    
dic_dates
```

```python
#Q9: For each country in the dataset, 
#when is the last date for which we have information about the number of fully vaccinated people?
dict_country_fully = {}

for i in range(len(csv_data)):
    country = cell(i,"country")
    date = cell(i,"date")
    people_fully_vaccinated = cell(i,"people_fully_vaccinated")
    last_date = ""
            
    # if the country is not a key to the dictionary, add it and set it's value to None
    if country not in dict_country_fully:
        dict_country_fully[country] = None
    # If the country is in the dictionary and the current date is > than the one stored we should update it 
    if people_fully_vaccinated != None: 
        dict_country_fully[country] = date
dict_country_fully
```

```python
def most_recent_total(col_name, given_date):
    '''return a dictionary mapping each country to the most recent column value in the data 
    available by the given date; if no data is available, the value is None.'''
    country_info = {}
    for row_idx in range(len(csv_data)):
        country = cell(row_idx,"country")
        date = cell(row_idx,"date")
        col_value = cell(row_idx,col_name)
        if country not in country_info:
            country_info[country] = None
        # Make sure the column value is not missing AND Check if the date is on or before given_date:
        if col_value != None and date <= given_date:
            country_info[country] = col_value
    return country_info
```

```python
#Q10: How many people were estimated to be fully vaccinated in each country, as of 31 Jan 2022?
dict_vacc_each = most_recent_total("people_fully_vaccinated", "01/31/2022")
dict_vacc_each
```

```python
#Q11: Among countries with population less than 100 million, which country is estimated to have had the highest 
#Â number of fully vaccinated people by 31 Jan 2022?
lista_paises = list()

for rows in populations:
    if populations[rows] != None and populations[rows] < 100000000:
        lista_paises.append(rows)
lista_paises = list(set(lista_paises))

lista_paises
max_value = 0
max_key = None

# iterate over the dict to find the max
for row in dict_vacc_each:
    if row not in lista_paises: 
        continue
    curr_value = dict_vacc_each[row]
    if curr_value == None:
            continue
    if curr_value > max_value or max_key == None:
        max_value = curr_value
        max_key = row
max_key
```

```python
vaccination_stats = {}

for date in dates:
    daily_vax_dict = daily_vaccinations_on(date) # a dict mapping each country to number of daily_vaccinations on date
    total_vax_dict = most_recent_total("total_vaccinations", date) # a dict mapping each country to the most recent number of total_vaccinations on date
    people_vax_dict = most_recent_total("people_vaccinated", date) # a dict mapping each country to the most recent number of people_vaccinated on date
    fully_vax_dict = most_recent_total("people_fully_vaccinated", date) # a dict mapping each country to most recent number of people_fully_vaccinated on date
    total_booster_dict = most_recent_total("total_boosters", date) # a dict mapping each country to most recent number of total_boosters on date
    for country in names:
        if country not in vaccination_stats:
            vaccination_stats[country] = {}
        vaccination_stats[country][date] = {}
        vaccination_stats[country][date]['country'] = country
        vaccination_stats[country][date]['date'] = date
        vaccination_stats[country][date]['daily_vaccinations'] = daily_vax_dict[country]
        vaccination_stats[country][date]['total_vaccinations'] = total_vax_dict[country]
        vaccination_stats[country][date]['people_vaccinated'] = people_vax_dict[country]
        vaccination_stats[country][date]['people_fully_vaccinated'] = fully_vax_dict[country]
        vaccination_stats[country][date]['total_boosters'] = total_booster_dict[country]
        vaccination_stats[country][date]['population'] = populations[country]
        
```

```python
#Q16: What is the estimated ratio of people who have been partly vaccinated to people who have been fully 
#vaccinated in each country, as of Jan 31, 2022?
dict_ratio1 = {}
for country in vaccination_stats:
    count_partial = vaccination_stats[country]['01/31/2022']['people_vaccinated']
    count_fully = vaccination_stats[country]['01/31/2022']['people_fully_vaccinated']
    
    if count_partial == None or count_fully == None:
        dict_ratio1[country] = None
    else: 
        dict_ratio1[country] = count_partial/count_fully - 1
dict_ratio1   
```

```python
#Q17: Which country has administered the most number of booster shots as of Jan 31, 2022
dict_boosters = {}
for country in vaccination_stats:
    dict_boosters[country] = vaccination_stats[country]['01/31/2022']['total_boosters']
get_max_key(dict_boosters)
```

```python
#Q18: Which country is estimated to have administered the most number of booster shots per capita, 
#as of 25 Jan 2022?
dict_boosters_cap = {}
for country in vaccination_stats:
    total_boosters = vaccination_stats[country]['01/25/2022']['total_boosters']
    population = vaccination_stats[country]['01/25/2022']['population']
    if total_boosters == None or population == None:
        dict_boosters_cap[country] = None
    else: 
        boost_cap = total_boosters/population
        dict_boosters_cap[country] = boost_cap
get_max_key(dict_boosters_cap)
```

```python
#Q19: What is the percentage of population boosted by Jan 31, 2022 of each country in the dataset?
dict_boosters_per = {}
for country in vaccination_stats:
    total_boosters = vaccination_stats[country]['01/31/2022']['total_boosters']
    population = vaccination_stats[country]['01/31/2022']['population']
    
    if total_boosters == None or population == None:
        continue
    else: 
        boost_cap = total_boosters/population * 100
        dict_boosters_per[country] = boost_cap
dict_boosters_per
```

```python
#Q20: Among all the countries in the dataset, what is the rank of Canada's percentage of population boosted?
ranking = 0

for country in dict_boosters_per:
    if dict_boosters_per["Canada"] < dict_boosters_per[country]:
        ranking += 1
ranking += 1
ranking
```
